```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
H = 50
Vs = 220
g = 9.81
gamma = 18
ro = gamma / g
G = Vs**2 * ro

omega = np.arange(0, 60.1, 0.1)
omega_2 = omega**2

Beta = [0, 0.03]

for beta in Beta:
    Eta = 2 * beta * G / omega
    P = np.sqrt(ro * omega_2 / (G + 1j * Eta * omega))
    A_om = 2 / abs((np.exp(1j * P * H) + np.exp(1j * -1 * P * H)))
    
    if beta == 0:
        y1 = A_om
    elif beta == 0.03:
        y2 = A_om

i = omega / 2 / np.pi
plt.plot(i, y1, 'b--', i, y2, 'green')
plt.xlabel('Frequency (HZ)')
plt.ylabel('Amplification ratio')
plt.legend(['Beta=0', 'Beta=0.03'])
plt.xlim([0, 10])
plt.ylim([0, 30])
plt.title('Rigid Rock')
plt.grid(True)
plt.show()
```
